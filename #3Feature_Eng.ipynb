A tweet is written in different emotions, vaderSentiment API is used to do the sentiment analysis of the tweet. 

It returns 1, 0, -1 which corresponds to Positive tweet, Neutral tweet, Negative tweet respectively

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyzer_scores(text):
    score = analyser.polarity_scores(text)
    lb = score['compound']
    if lb >= 0.05:
        return 1
    elif (lb > -0.05) and (lb < 0.05):
        return 0
    else:
        return -1
```

```python
tweet['text']=tweet['text'].map(sentiment_analyzer_scores, na_action='ignore')
```

```python
#Defining the virality classification function, here if the tweet is set to viral if the tweet has 1000 or more number of retweets
def ret(mm):
    mm=int(mm)
    if mm>1000:
        return 1
    else:
        return 0
```

```python
tweet['retweets']=tweet['retweets'].map(ret, na_action='ignore')
```

```python
from datetime import datetime
```

```python
#Converting timestamp into different categories such as Morning, Noon, Evening, Night
def time_of_day(m):
    datetime_object = datetime.strptime(m, '%Y-%m-%d %H:%M:%S')
    d = int(datetime_object.strftime("%H"))
    if d<12:
        return 'Morning'
    elif d>=12 and d<16:
        return 'Noon'
    elif d>=16 and d<21:
        return 'Evening'
    else:
        return 'Night'
    
```

```python
twit['timestamp']=twit['timestamp'].map(time_of_day, na_action='ignore')
```
